```python
def manage_parking(capacity_limit, logs):
    vehicles_inside = 0
    highest_count = 0
    for record in logs:
        if record == "IN":
            vehicles_inside += 1
        elif record == "OUT":
            if vehicles_inside > 0:
                vehicles_inside -= 1
        if vehicles_inside > highest_count:
            highest_count = vehicles_inside
    print("Currently Parked Vehicles:", vehicles_inside)
    print("Peak Parking Usage:", highest_count)
    if vehicles_inside > capacity_limit:
        print("Parking Status: Capacity Exceeded - Alert!")
    else:
        print("Parking Status: Available")
parking_capacity = 50
vehicle_activity = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(parking_capacity, vehicle_activity)
```

```python
def delivery_time_estimator(distance_km, traffic_level, weather_condition):
    minutes_per_km = 60 / 30
    travel_time = distance_km * minutes_per_km
    if traffic_level == "Low":
        travel_time += 3
    elif traffic_level == "Medium":
        travel_time += 8
    elif traffic_level == "High":
        travel_time += 15
    if weather_condition == "Clear":
        travel_time += 0
    elif weather_condition == "Rainy":
        travel_time += 10
    elif weather_condition == "Stormy":
        travel_time += 20
    print("Estimated Delivery Time:", int(travel_time), "minutes")
distance = 8
traffic = "High"
weather = "Rainy"
delivery_time_estimator(distance, traffic, weather)
```

```python
def theatre_occupancy_checker(total_seats, seat_data):
    booked_seats = sum(seat_data)
    occupancy = (booked_seats / total_seats) * 100
    print("Occupancy:", round(occupancy), "%")
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"
    print("Show Status:", status)
    if occupancy >= 90:
        print("Recommendation: Add an extra show due to high demand")
total_seats = 200
booked_seats_list = [1] * 150
theatre_occupancy_checker(total_seats, booked_seats_list)
```

```python
def server_load_checker(cpu_values):
    total = 0
    count = 0
    for reading in cpu_values:
        total += reading
        count += 1
    average = total / count
    print("Average CPU Load:", int(average), "%")
    if average < 50:
        condition = "Normal"
    elif 50 <= average <= 80:
        condition = "Warning"
    else:
        condition = "Critical"
    print("Server Status:", condition)
cpu_readings = [45, 60, 70, 85, 90]
server_load_checker(cpu_readings)
```

```python
def monitor_classroom_usage(usage_info):
    max_allowed_hours = 8
    excessive_usage = []
    for item in usage_info:
        hours_used = usage_info[item]
        if hours_used > max_allowed_hours:
            excessive_usage.append(item)
    if len(excessive_usage) > 0:
        print("Overused Resources:", ", ".join(excessive_usage))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_classroom_usage(resources)
```

```python
def manage_event_capacity(max_capacity, incoming_registrations):
    confirmed_count = 0
    waiting_count = 0
    if incoming_registrations <= max_capacity:
        confirmed_count = incoming_registrations
        registration_state = "Open"
    else:
        confirmed_count = max_capacity
        waiting_count = incoming_registrations - max_capacity
        registration_state = "Closed"
    print("Confirmed Registrations:", confirmed_count)
    print("Waitlisted Users:", waiting_count)
    print("Registration Status:", registration_state)
event_capacity = 100
registrations = 105
manage_event_capacity(event_capacity, registrations)
```
