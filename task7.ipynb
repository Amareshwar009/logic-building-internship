```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default state is OFF
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(f"{self.name} is {self.status}")
def main():
    light_name = input("Enter Light Name: ")
    action = input("Enter Action (ON/OFF): ").strip().upper()
    light = SmartLight(light_name)
    if action == "ON":
        light.turn_on()
        light.display_status()
    elif action == "OFF":
        light.turn_off()
        light.display_status()
    else:
        print("Invalid Action! Please enter ON or OFF.")
if __name__ == "__main__":
    main()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("------------------")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
def main():
    name = input("Employee Name: ")
    emp_id = input("Employee ID: ")
    department = input("Department: ")
    employee = Employee(name, emp_id, department)
    employee.display_id_card()
if __name__ == "__main__":
    main()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("\nContact Saved")
        print("--------------")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
def main():
    name = input("Contact Name: ")
    phone = input("Phone Number: ")
    contact = Contact(name, phone)
    contact.display_contact()
if __name__ == "__main__":
    main()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def print_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))
def main():
    product_name = input("Product Name: ")
    price = input("Price: ")
    product = Product(product_name, price)
    product.print_price_tag()
if __name__ == "__main__":
    main()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_movie_details(self):
        print("\nMovie:", self.name)
        print(f"Rating: {self.rating} / 5")
def main():
    movie_name = input("Movie Name: ")
    rating = input("Rating: ")
    movie = Movie(movie_name, rating)
    movie.display_movie_details()
if __name__ == "__main__":
    main()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    def print_delivery_details(self):
        print("\nDelivery Details")
        print("-----------------")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")
def main():
    name = input("Customer Name: ")
    address = input("Address: ")
    delivery = Delivery(name, address)
    delivery.print_delivery_details()
if __name__ == "__main__":
    main()
```
