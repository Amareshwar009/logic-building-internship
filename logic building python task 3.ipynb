```python
employees={
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "Suresh":85
}
highest_score=0
for score in employees.values():
    if score>highest_score:
        highest_score=score
eligible_employees=[]
for name, score in employees.items():
    if score==highest_score:
        eligible_employees.append(name)
print("Top Performers Eligible for Bonus:", end="")
print(",".join(eligible_employees)+f"(score:{highest_score})")
```

```python
search_query="buy mobile phone buy phone online"
search_query=search_query.lower()
punctuation_marks=".!?;:'\"()-"
for symbol in punctuation_marks:
    search_query=search_query.replace(symbol,"")
word_list=search_query.split()
frequency={}
for word in word_list:
    frequency[word]=frequency.get(word,0)+1
multiple_searches={}
for word, count in frequency.items():
    if count>1:
        multiple_searches[word]=count
print(multiple_searches)
```

```python
sensor_readings=[3,4,7,8,10,12,5]
validated_data=[]
for hour_index, value in enumerate(sensor_readings):
    if value%2==0:
        validated_data.append((hour_index, value))
print("Valid Sensor Readings (Hour, Value):")
print(validated_data)
```

```python
email_ids = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_stats={}
for address in email_ids:
    parts =address.partition("@")
    domain_name=parts[2]
    domain_stats[domain_name]=domain_stats.get(domain_name, 0)+1
user_total = len(email_ids)
for domain_name in domain_stats:
    usage_percent = (domain_stats[domain_name] / user_total) * 100
    print(f"{domain_name}: {round(usage_percent)}%")
```

```python
daily_sales=[1200,1500,900,2200,1400,3000]
count_days=len(daily_sales)
running_total=0
for value in daily_sales:
    running_total+=value
average_value=running_total/count_days
increase_limit=average_value+(average_value*0.30)
day_number=1
for value in daily_sales:
    if value>increase_limit:
        print("Day",day_number,":",value)
    day_number+=1
```
