```python
def check_recharge():
    available_plans = {199, 299, 399, 599}
    while True:
        user_input = input("Enter recharge amount: ₹")
        if not user_input.isdigit():
            print("Invalid input! Please enter numbers only.")
            continue
        amount = int(user_input)
        if amount < 50:
            print("Amount should be ₹50 or more.")
            continue
        if amount not in available_plans:
            print("Invalid plan. Available plans are:", available_plans)
            continue
        print("Recharge successful for ₹", amount)
        break
check_recharge()
```

```python
def inventory_alert(stock_data):
    for product in stock_data:
        quantity = stock_data[product]
        if quantity < 15:
            print(product + ": Reorder Alert (Current Stock =", quantity, ")")
        else:
            print(product + ": Stock OK (Current Stock =", quantity, ")")
products = {
    "Notebook": 20,
    "Pen": 12,
    "Pencil": 8,
    "Eraser": 25
}
inventory_alert(products)
```

```python
def student_result(marks):
    total_marks = 0
    for value in marks:
        total_marks = total_marks + value
    average_marks = total_marks / len(marks)
    if average_marks >= 50:
        status = "Pass"
    else:
        status = "Fail"
    return average_marks, status
marks_list = [65, 48, 72, 50, 40]
avg, result = student_result(marks_list)
print("Student Marks:", marks_list)
print("Average Marks:", avg)
print("Final Result:", result)
```

```python
def fare_calculator(distance, is_peak):
    base_amount = 50
    rate_per_km = 12
    total_fare = base_amount + (distance * rate_per_km)
    if is_peak == "yes":
        extra_charge = total_fare * 0.25
        total_fare = total_fare + extra_charge
    return total_fare
while True:
    user_distance = input("Enter travel distance (in km): ")
    if not user_distance.replace('.', '', 1).isdigit():
        print("Please enter a valid numeric distance.\n")
        continue
    distance_value = float(user_distance)
    if distance_value <= 0:
        print("Distance must be greater than zero.\n")
        continue
    peak_input = input("Is this a peak hour ride? (yes/no): ").lower()
    if peak_input not in ["yes", "no"]:
        print("Please type 'yes' or 'no' only.\n")
        continue
    final_amount = fare_calculator(distance_value, peak_input)
    print("Total Cab Fare: ₹", final_amount)
    break
```

```python
def attendance_checker(records):
    present_count = 0
    for day in records:
        if day == "P":
            present_count += 1
    total_days = len(records)
    if total_days == 0:
        return 0, "No Data Available"
    attendance_percent = (present_count / total_days) * 100
    if attendance_percent >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return attendance_percent, status
employee_attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
percentage, result = attendance_checker(employee_attendance)
print("Attendance Record:", employee_attendance)
print("Attendance Percentage:", percentage, "%")
print("Eligibility:", result)
```

```python
def password_checker(user_password):
    digit_found = False
    special_found = False
    if len(user_password) < 8:
        return "Weak Password (Minimum length must be 8 characters)"
    for character in user_password:
        if character >= '0' and character <= '9':
            digit_found = True
        if character == '@' or character == '#' or character == '$':
            special_found = True
    if digit_found and special_found:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"
password_a = "Secure1@"
password_b = "Password"
password_c = "abc12"
print("Password:", password_a)
print("Result:", password_checker(password_a))
print()
print("Password:", password_b)
print("Result:", password_checker(password_b))
print()
print("Password:", password_c)
print("Result:", password_checker(password_c))
```
