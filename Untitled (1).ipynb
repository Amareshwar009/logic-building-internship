```python
scores=[45,78,90,33,60]
passed_students=0
failed_students=0
index=0
while index<len(scores):
    current_scores=scores[index]
    if current_scores>=50:
        passed_students=passed_students+1
    else:
        failed_students=failed_students+1
    index+=1
print("result summary")
print("")
print("number of students:",len(scores))
print("passed:", passed_students)
print("failed:", failed_students)
```

```python
raw_names=["alice","bob","charlie"]
cleaned_names=[]
position=0
while position<len(raw_names):
    name_value=raw_names[position]
    formatted_name=name_value.strip().lower()
    cleaned_names.append(formatted_name)
    position+=1
print("cleaned names list:")
print(cleaned_names)
```
